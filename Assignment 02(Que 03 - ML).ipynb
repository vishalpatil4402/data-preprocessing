## Preprocessing

```python
data.select_dtypes('object').columns
```

```python
data.Class.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
le = LabelEncoder()
```

```python
Class = le.fit_transform(data.Class)
```

```python
ohe = OneHotEncoder()
```

```python
ohe_Class = ohe.fit_transform(Class.reshape(-1,1))
```

```python
le.classes_
```

```python
Class_df = pd.DataFrame(ohe_Class.toarray(), dtype=int,columns=le.classes_)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
data.columns
```

```python
df = data[['YearStart','Data_Value', 'Sample_Size']]
```

```python
df.plot.kde()
```

```python
mm = MinMaxScaler()
```

```python
scaled_np = mm.fit_transform(df)
```

```python
Class_np = Class_df.values
```

```python
df = data[['Data_Value_Alt','Low_Confidence_Limit','LocationID','Sample_Size','YearEnd']]
```

```python
data_np = df.values
```

```python
feature_data = np.hstack([data_np, scaled_np, Class_np])
```

```python
target_data = data.Sample_Size
```

```python
feature_data.shape
```

```python
target_data.value_counts()
```

## Model Building

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier, PassiveAggressiveClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = [ LogisticRegression(class_weight='balanced'), SGDClassifier(max_iter=10), PassiveAggressiveClassifier(max_iter=20), RandomForestClassifier(n_estimators=20)]
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainX,testX,trainY,testY = train_test_split(feature_data,target_data)
```

```python
for model in models:
    model.fit(trainX,trainY)
    print (model.score(testX,testY))
```
